```
# Maria Eduarda - Gerenciamento de Usuários
# Login/Logout utilizando matriz

class SistemaUsuarios:
    def __init__(self):
        self.usuarios = []  # Matriz para login/logout
        self.usuario_logado = []  # Pilha para usuário logado

    def cadastrar_usuario(self, nome, email, senha):
        self.usuarios.append([nome, email, senha])
        print(f"Usuário {nome} cadastrado com sucesso!")

    def login(self, email, senha):
        for usuario in self.usuarios:
            if usuario[1] == email and usuario[2] == senha:
                self.usuario_logado.append(usuario)
                print(f"Usuário {usuario[0]} logado com sucesso!")
                return True
        print("Login falhou. Verifique suas credenciais.")
        return False

    def logout(self):
        if self.usuario_logado:
            usuario = self.usuario_logado.pop()
            print(f"Usuário {usuario[0]} deslogado com sucesso!")
        else:
            print("Nenhum usuário logado.")

# Teste individual
if __name__ == "__main__":
    sistema = SistemaUsuarios()
    sistema.cadastrar_usuario("Carlos", "carlos@email.com", "1234")
    sistema.login("carlos@email.com", "1234")
    sistema.logout()


# Ana Clara - Sistema de Produtos e Pagamento
# Lista de produtos e vetor de métodos de pagamento

class SistemaProdutos:
    def __init__(self):
        self.produtos = []  # Lista de produtos
        self.metodos_pagamento = []  # Vetor de métodos de pagamento

    def adicionar_produto(self, produto):
        self.produtos.append(produto)
        print(f"Produto '{produto}' adicionado à lista.")

    def listar_produtos(self):
        print("Produtos disponíveis:")
        for produto in self.produtos:
            print(f"- {produto}")

    def escolher_metodo_pagamento(self, metodo):
        self.metodos_pagamento.append(metodo)
        print(f"Método de pagamento '{metodo}' registrado.")

# Teste individual
if __name__ == "__main__":
    sistema = SistemaProdutos()
    sistema.adicionar_produto("Passagem para Paris")
    sistema.adicionar_produto("Pacote de Hotel")
    sistema.listar_produtos()
    sistema.escolher_metodo_pagamento("Cartão de Crédito")


# Arthur - Carrinho de Compras e Finalização
# Utilizando fila para gerenciar o carrinho

class CarrinhoCompras:
    def __init__(self):
        self.carrinho = []  # Fila para carrinho de compras

    def adicionar_ao_carrinho(self, produto):
        self.carrinho.append(produto)
        print(f"Produto '{produto}' adicionado ao carrinho.")

    def finalizar_compra(self, usuario_logado):
        if usuario_logado and self.carrinho:
            print(f"Compra finalizada com sucesso! Itens adquiridos: {', '.join(self.carrinho)}")
            self.carrinho.clear()
        else:
            print("Erro ao finalizar compra. Verifique se está logado e há itens no carrinho.")

# Teste individual
if __name__ == "__main__":
    carrinho = CarrinhoCompras()
    carrinho.adicionar_ao_carrinho("Passagem para Paris")
    carrinho.finalizar_compra(True)  # Simulando um usuário logado
```
